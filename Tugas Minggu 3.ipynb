```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "n_movies.csv"
data = pd.read_csv(path)
data 
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
data.isna().sum()
```

```python
data.info()
```

```python
filtered_movie = data[['title', 'rating', 'description', 'votes']]
filtered_movie.head(10)
# sort by votes
filtered_movie = filtered_movie.sort_values(by='rating', ascending=False)
plot_filtered_movie = filtered_movie.head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='rating', y='title', data=plot_filtered_movie, palette='Blues')

for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', padding=3)

plt.title('Film/Serial dengan Rating Tertinggi di Netflix ', fontsize=16, fontweight='bold')
plt.xlabel('Rating', fontsize=12,fontweight='bold')
plt.ylabel('Film/Serial', fontsize=12,fontweight='bold')

# Menampilkan plot
plt.tight_layout()
plt.show()
```
